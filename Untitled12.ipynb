```python
import os

# Create main folder and subfolders
if not os.path.exists("lab_data"):
    os.makedirs("lab_data")
    print("lab_data created ")
else:
    print("lab_data already exists")

for folder in ["input", "output", "logs"]:
    path = os.path.join("lab_data", folder)
    if not os.path.exists(path):
        os.makedirs(path)

# List all files and folders
for item in os.listdir("lab_data"):
    full_path = os.path.join("lab_data", item)
    if os.path.isdir(full_path):
        print(f"{item} --> Directory ")
    else:
        print(f"{item} --> File ")

# Rename logs to history
logs_path = os.path.join("lab_data", "logs")
history_path = os.path.join("lab_data", "history")
if os.path.exists(logs_path):
    os.rename(logs_path, history_path)
    print("Renamed logs to history ")
```

```python
import os, shutil

# Create notes.txt
notes_path = os.path.join("lab_data", "input", "notes.txt")
with open(notes_path, "w") as f:
    f.write("Line 1: Python is a great programming language.\n")
    f.write("Line 2: It is easy to learn and use.\n")
    f.write("Line 3: Python supports many libraries.\n")
    f.write("Line 4: It is used in AI, web, and data science.\n")
    f.write("Line 5: I enjoy learning Python!\n")


with open(notes_path, "r") as f:
    lines = f.readlines()
    print("Total lines:", len(lines))


summary_path = os.path.join("lab_data", "output", "summary.txt")
shutil.copy(notes_path, summary_path)
print("Copied to summary.txt ")
```
